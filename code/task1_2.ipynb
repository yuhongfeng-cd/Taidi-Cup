```python
table = pd.pivot_table(city_data, values=['新增死亡', '新增治愈', '新增确诊', '累积死亡', '累积治愈', '累积确诊'], 
                      index=['城市', '日期'], 
                      aggfunc={'新增死亡': np.sum, 
                               '新增治愈': np.sum, 
                               '新增确诊': np.sum,
                               '累积死亡': np.sum,
                               '累积治愈': np.sum,
                               '累积确诊': np.sum,})
table
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import warnings
import re
warnings.filterwarnings('ignore')
```

```python
international_outbreak_data = pd.read_excel('../data/附件1.xlsx', sheet_name='国际疫情')
international_outbreak_data.head()
```

```python
international_outbreak_data['住院人数'] = international_outbreak_data['累计确诊'].values - international_outbreak_data['累计治愈'].values - international_outbreak_data['累计死亡'].values
international_outbreak_data.head()
```

```python
table = pd.pivot_table(international_outbreak_data, values=['住院人数'], 
                      index=['国家', '日期'])
table
```

```python
final_data = table.reset_index()
```

```python
yindu_data = final_data[final_data['国家'] == '印度']
yilang_data = final_data[final_data['国家'] == '伊朗']
yidali_data = final_data[final_data['国家'] == '意大利']
jianada_data = final_data[final_data['国家'] == '加拿大']
bilu_data = final_data[final_data['国家'] == '秘鲁']
nanfei_data = final_data[final_data['国家'] == '南非']
yindu_data.head()
```

```python
import pyecharts.options as opts
from pyecharts.charts import Line
from sklearn.cluster import KMeans

def result(X):
    X_input = np.array(X['住院人数']).reshape([len(np.array(X['住院人数'])), 1])
    kmeans = KMeans(n_clusters=2, n_jobs=1, random_state=0).fit(X_input)
    a, b = list(kmeans.cluster_centers_)[0], list(kmeans.cluster_centers_)[1]
    c = (
        Line()
        .add_xaxis(list(X['日期'].iloc[::5]))
        .add_yaxis("住院人数", list(X['住院人数'].iloc[::5]), is_smooth=True, label_opts=opts.LabelOpts(is_show=False))
#         .add_yaxis("", [min(a, b)[0] for _ in range(len(X))], is_smooth=True, label_opts=opts.LabelOpts(is_show=False))
#         .add_yaxis("", [max(a, b)[0] for _ in range(len(X))], is_smooth=True, label_opts=opts.LabelOpts(is_show=False))
        .add_yaxis("", min(a, b), is_smooth=True, label_opts=opts.LabelOpts(is_show=False))
        .add_yaxis("", max(a, b), is_smooth=True, label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(title_opts=opts.TitleOpts(title="南非现存患者趋势图"))
        
    )
    return kmeans.cluster_centers_, c
```

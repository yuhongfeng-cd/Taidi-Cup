```python
from sklearn.cluster import KMeans
```

```python
data_6 = {
#     'id':id_6,
    'x':hlist_6,
    'y':zlist_6
}
df_6 = pd.DataFrame(data_6)
```

```python
data_10 = {
    'x':hlist_10,
    'y':zlist_10
}
df_10 = pd.DataFrame(data_10)
```

```python
k = 2
kmeans_model = KMeans(n_clusters = k,n_jobs = 2,random_state=123)
fit_kmeans = kmeans_model.fit(df_6)
kmeans_model.cluster_centers_
```

```python
count6=list(kmeans_model.labels_)
```

```python
from pyecharts import options as opts
from pyecharts.charts import Bar
from pyecharts.commons.utils import JsCode
from pyecharts.globals import ThemeType

list2 = [
    {"value": 68, "percent": 68 / (68 + 13)},
    {"value": 93, "percent": 93 / (19 + 93)}
]

list3 = [
    {"value": 13, "percent": 13 / (13 + 68)},
    {"value": 19, "percent": 19 / (93 + 19)}
]

c = (
    Bar(init_opts=opts.InitOpts(theme=ThemeType.LIGHT))
    .add_xaxis(['前6天', '前10天'])
    .add_yaxis("高危险区域", list2, stack="stack1", category_gap="50%")
    .add_yaxis("危险区域", list3, stack="stack1", category_gap="50%")
    .set_series_opts(
        label_opts=opts.LabelOpts(
            position="right",
            formatter=JsCode(
                "function(x){return Number(x.data.percent * 100).toFixed() + '%';}"
            ),
        )
    )
)
c.render_notebook()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import warnings
import re
warnings.filterwarnings('ignore')
```

```python
provinces_data = pd.read_csv("../results/result1/task1_2.csv", encoding='gbk')
provinces_data.head()
```

```python
tmp = provinces_data[['省份', '日期', '新增确诊人数']]
tmp.head()
```

```python
tmp['新增确诊人数'] = provinces_data['累积确诊人数'].values - provinces_data['累积治愈人数'].values - provinces_data['累积死亡人数'].values
tmp.head()
```

```python
final_data = tmp
final_data.columns = ['省份', '日期', '住院人数']
final_data.to_csv("./task1_3.csv", encoding='gbk', index=False)
```

```python
hubei_data = final_data[final_data['省份']=='湖北']
dates = []
for i in range(1, 7):
    dates.append('2020/'+ str(i) +'/15')
# print(dates)
last_hubei_data = pd.DataFrame()
for d in dates:
    last_hubei_data = last_hubei_data.append(hubei_data[hubei_data['日期']==d], ignore_index = True)
last_hubei_data
```
